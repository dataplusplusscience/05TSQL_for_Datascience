### Monthly Sales Trend:

```python

# execute a query
cursor.execute('''SELECT
 DATEPART(MONTH, OrderDate) AS Month,
 SUM(TotalDue) AS TotalSales
FROM
 SalesLT.SalesOrderHeader
WHERE
 DATEPART(YEAR, OrderDate) = 2008
GROUP BY
 DATEPART(MONTH, OrderDate)
ORDER BY
 Month;''')

# Fetch all rows from the executed query
rows = cursor.fetchall()

# Get the column names
columns = [column[0] for column in cursor.description]

# Convert the rows into a list of dictionaries
data = [dict(zip(columns, row)) for row in rows]

# Create a DataFrame from the list of dictionaries
monthly_sales_trend = pd.DataFrame(data)
monthly_sales_trend.head()
```

### Customer Segment Analysis

```python

# execute a query
cursor.execute('''SELECT
 C.CustomerID,
 COUNT(O.SalesOrderID) AS OrderCount,
 SUM(O.TotalDue) AS TotalRevenue
FROM
 SalesLT.SalesOrderHeader O
JOIN
 SalesLT.Customer C
 ON O.CustomerID = C.CustomerID
GROUP BY
 C.CustomerID;''')

# Fetch all rows from the executed query
rows = cursor.fetchall()

# Get the column names
columns = [column[0] for column in cursor.description]

# Convert the rows into a list of dictionaries
data = [dict(zip(columns, row)) for row in rows]

# Create a DataFrame from the list of dictionaries
cust_segment_analysis = pd.DataFrame(data)
cust_segment_analysis.head()
```

### Best Selling Products

```python

# execute a query
cursor.execute('''SELECT TOP 10
 P.Name AS ProductName,
 SUM(D.LineTotal) AS TotalRevenue,
 SUM(D.OrderQty) AS TotalQuantitySold
FROM
 SalesLT.SalesOrderDetail D
JOIN
 SalesLT.Product P
 ON D.ProductID = P.ProductID
GROUP BY
 P.Name
ORDER BY
 TotalRevenue DESC;;''')

# Fetch all rows from the executed query
rows = cursor.fetchall()

# Get the column names
columns = [column[0] for column in cursor.description]

# Convert the rows into a list of dictionaries
data = [dict(zip(columns, row)) for row in rows]

# Create a DataFrame from the list of dictionaries
cust_segment_analysis = pd.DataFrame(data)
cust_segment_analysis.head()
```

### Repeat Customers

```python

# execute a query
cursor.execute('''SELECT
 C.FirstName + C.LastName AS CustomerName,
 COUNT(O.SalesOrderID) AS TotalOrders,
 SUM(O.TotalDue) AS TotalSpent
FROM
 SalesLT.SalesOrderHeader O
JOIN
 SalesLT.Customer C
 ON O.CustomerID = C.CustomerID
GROUP BY
 C.FirstName, C.LastName
HAVING
 COUNT(O.SalesOrderID) > 1
ORDER BY
 TotalSpent DESC;''')

# Fetch all rows from the executed query
rows = cursor.fetchall()

# Get the column names
columns = [column[0] for column in cursor.description]

# Convert the rows into a list of dictionaries
data = [dict(zip(columns, row)) for row in rows]

# Create a DataFrame from the list of dictionaries
cust_segment_analysis = pd.DataFrame(data)
cust_segment_analysis.head()
```

### Age Group Analysis

```python



# execute a query
cursor.execute('''SELECT
 CASE
     WHEN Age >= 18 AND Age <= 30 THEN '18â€“30'
     WHEN Age > 30 AND Age <= 45 THEN '31â€“45'
     WHEN Age > 45 AND Age <= 60 THEN '46â€“60'
 ELSE 'Above 60'
 END AS AgeGroup,
 COUNT(*) AS TotalCustomers,
 SUM(O.TotalDue) AS TotalRevenue
FROM
 (
 SELECT
 CustomerID,
 DATEDIFF(YEAR, BirthDate, GETDATE()) AS Age
 FROM
 SalesLT.Customer
 ) C
JOIN
 SalesLT.SalesOrderHeader O
 ON C.CustomerID = O.CustomerID
GROUP BY
 AgeGroup
ORDER BY
 AgeGroup;''')

# Fetch all rows from the executed query
rows = cursor.fetchall()

# Get the column names
columns = [column[0] for column in cursor.description]

# Convert the rows into a list of dictionaries
data = [dict(zip(columns, row)) for row in rows]

# Create a DataFrame from the list of dictionaries
cust_segment_analysis = pd.DataFrame(data)
cust_segment_analysis.head()
```
